```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
```

```python
indoor = pd.read_csv('indoor_data.csv')
outside = pd.read_csv('outside_data.csv')

indoor = indoor.iloc[30:301]
outside = outside.iloc[30:301]
```

```python
print("Indoor Mean and Standard Deviation")
mean_T = np.mean(indoor['Temperature'])
std_T = np.std(indoor['Temperature'])
varianceT = np.var(indoor['Temperature'])
uncertaintyT_indoor = (std_T/(270)**.5)
print("Temperature", mean_T, std_T, varianceT, uncertaintyT_indoor)

mean_H = np.mean(indoor['Humidity'])
std_H = np.std(indoor['Humidity'])
varianceH = np.var(indoor['Humidity'])
uncertaintyH_indoor = (std_H/(270)**.5)
print("Humidity", mean_H, std_H, varianceH, uncertaintyH_indoor)

mean_P = np.mean(indoor['Pressure'])
std_P = np.std(indoor['Pressure'])
varianceP = np.var(indoor['Pressure'])
uncertaintyP_indoor = (std_P/(270)**.5)
print("Pressure", mean_P, std_P, varianceP, uncertaintyP_indoor)

mean_A = np.mean(indoor['Altitude'])
std_A = np.std(indoor['Altitude'])
varianceA = np.var(indoor['Altitude'])
uncertaintyA_indoor = (std_A/(270)**.5)
print("Altitude", mean_A, std_A, varianceA, uncertaintyA_indoor)
```

```python
print("Outside Mean, Standard Deviation, Variance, and Uncertainty")
mean_T = np.mean(outside['Temperature'])
std_T = np.std(outside['Temperature'])
varianceT = np.var(outside['Temperature'])
uncertaintyT_outside = (std_T/(270)**.5)
print("Temperature", mean_T, std_T,uncertaintyT_outside, varianceT)

mean_H = np.mean(outside['Humidity'])
std_H = np.std(outside['Humidity'])
varianceH = np.var(outside['Humidity'])
uncertaintyH_outside = (std_H/(270)**.5)
print("Humidity", mean_H, std_H, uncertaintyH_outside, varianceH)

mean_P = np.mean(outside['Pressure'])
std_P = np.std(outside['Pressure'])
varianceP = np.var(outside['Pressure'])
uncertaintyP_outside = (std_P/(270)**.5)
print("Pressure", mean_P, std_P,uncertaintyP_outside, varianceP)

mean_A = np.mean(outside['Altitude'])
std_A = np.std(outside['Altitude'])
varianceA = np.var(outside['Altitude'])
uncertaintyA_outside = (std_A/(270)**.5)
print("Altitude", mean_A, std_A, uncertaintyA_outside, varianceA)
```

```python
plt.plot(indoor['Time'], indoor['Temperature'], linestyle = 'solid', color='red')
plt.plot(indoor['Time'], indoor['Humidity'], linestyle = 'solid', color='blue')
plt.plot(indoor['Time'], indoor['Pressure'], linestyle = 'solid', color='green')
plt.plot(indoor['Time'], indoor['Altitude'], linestyle = 'solid', color='purple')

plt.xlim(30, 300)
plt.ylim(0, 100)
plt.title('Indoor Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Temperature', 'Humidity', 'Altitude'])

plt.show()

plt.plot(indoor['Time'], indoor['Pressure'], linestyle = 'solid', color='green')
plt.xlim(30, 300)
plt.ylim(950, 1050)
plt.title('Indoor Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Pressure'])

plt.show()
```
