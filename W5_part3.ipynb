```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python
indoor = pd.read_csv('indoor_data.csv')
outside = pd.read_csv('outside_data.csv')
```

```python
plt.plot(indoor['Time'], indoor['Temperature'], linestyle = 'solid', color='red')
plt.plot(indoor['Time'], indoor['Humidity'], linestyle = 'solid', color='blue')
plt.plot(indoor['Time'], indoor['Pressure'], linestyle = 'solid', color='green')
plt.plot(indoor['Time'], indoor['Altitude'], linestyle = 'solid', color='purple')

plt.xlim(30, 300)
plt.ylim(0, 100)
plt.title('Indoor Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Temperature', 'Humidity', 'Altitude'])

plt.show()

plt.plot(indoor['Time'], indoor['Pressure'], linestyle = 'solid', color='green')
plt.xlim(30, 300)
plt.ylim(950, 1050)
plt.title('Indoor Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Pressure'])

plt.show()
```

```python
plt.plot(outside['Time'], outside['Temperature'], linestyle = 'dashed', color='red')
plt.plot(outside['Time'], outside['Humidity'], linestyle = 'dashed', color='blue')
plt.plot(outside['Time'], outside['Altitude'], linestyle = 'dashed', color='purple')
# title and axis labels
plt.xlim(30, 300)
plt.ylim(0, 100)
plt.title('Outside Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Temperature', 'Humidity', 'Altitude'])

plt.show()

plt.plot(outside['Time'], outside['Pressure'], linestyle = 'dashed', color='green')
plt.xlim(30, 300)
plt.ylim(950, 1050)
plt.title('Outside Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Pressure'])

plt.show()
```

```python
plt.plot(indoor['Time'], indoor['Temperature'], linestyle = 'solid', color='red')
plt.plot(indoor['Time'], indoor['Humidity'], linestyle = 'solid', color='blue')
plt.plot(indoor['Time'], indoor['Pressure'], linestyle = 'solid', color='green')
plt.plot(indoor['Time'], indoor['Altitude'], linestyle = 'solid', color='purple')


plt.plot(outside['Time'], outside['Temperature'], linestyle = 'dashed', color='red')
plt.plot(outside['Time'], outside['Humidity'], linestyle = 'dashed', color='blue')
plt.plot(outside['Time'], outside['Pressure'], linestyle = 'dashed', color='green')
plt.plot(outside['Time'], outside['Altitude'], linestyle = 'dashed', color='purple')

plt.xlim(30, 300)
plt.ylim(0, 1100)
plt.title('Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Temperature', 'Humidity', 'Pressure', 'Altitude'])

plt.show()
```

```python
plt.hist(indoor["Temperature"],bins=10)
plt.hist(outside["Temperature"],bins=10)

plt.hist(indoor["Humidity"],bins=10)
plt.hist(outside["Humidity"],bins=10)

plt.hist(indoor["Altitude"],bins=70)
plt.hist(outside["Altitude"],bins=70)

plt.xlim(5, 120)
plt.ylim(0, 500)
plt.legend(['T Indoor', 'T Outside', 'H Indoor', 'H Outside', 'A Indoor', 'A Outside'])
plt.show()

plt.hist(indoor["Pressure"],bins=5)
plt.hist(outside["Pressure"],bins=5)
plt.xlim(900, 1050)
plt.ylim(0, 500)
plt.title('Pressure Histogram (Indoor and Outdoor)')
plt.show() 
```
