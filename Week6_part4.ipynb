```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
import scipy 
```

```python
# Temp vs humidity 
construction = pd.read_csv('construction_data.csv')
construction = construction.iloc[120:421]
plt.plot(np.unique(construction['Temperature']), np.poly1d(np.polyfit(construction['Temperature'], construction['Humidity'] , 1))(np.unique(construction['Temperature'])))
plt.plot(construction['Temperature'], construction['Humidity'], linestyle = 'solid', color='red')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.legend(['Line of best fit', 'Temperature vs. Humidity'])
plt.show
```
