```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
```

```python
construction = pd.read_csv('construction_data.csv')

construction = construction.iloc[120:421]
```

```python
print("Construction Mean and Standard Deviation")
mean_T = np.mean(construction['Temperature'])
std_T = np.std(construction['Temperature'])
varianceT = np.var(construction['Temperature'])
uncertaintyT_construction = (std_T/(270)**.5)
print("Temperature", mean_T, std_T, varianceT, uncertaintyT_construction)

mean_H = np.mean(construction['Humidity'])
std_H = np.std(construction['Humidity'])
varianceH = np.var(construction['Humidity'])
uncertaintyH_construction = (std_H/(270)**.5)
print("Humidity", mean_H, std_H, varianceH, uncertaintyH_construction)

mean_P = np.mean(construction['Pressure'])
std_P = np.std(construction['Pressure'])
varianceP = np.var(construction['Pressure'])
uncertaintyP_construction = (std_P/(270)**.5)
print("Pressure", mean_P, std_P, varianceP, uncertaintyP_construction)

mean_A = np.mean(construction['Altitude'])
std_A = np.std(construction['Altitude'])
varianceA = np.var(construction['Altitude'])
uncertaintyA_construction = (std_A/(270)**.5)
print("Altitude", mean_A, std_A, varianceA, uncertaintyA_construction)
```

```python
plt.plot(construction['Time'], construction['Temperature'], linestyle = 'solid', color='red')
plt.plot(construction['Time'], construction['Humidity'], linestyle = 'solid', color='blue')
plt.plot(construction['Time'], construction['Pressure'], linestyle = 'solid', color='green')
plt.plot(construction['Time'], construction['Altitude'], linestyle = 'solid', color='purple')

plt.xlim(120, 420)
plt.ylim(0, 100)
plt.title('construction Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Temperature', 'Humidity', 'Altitude'])

plt.show()

plt.plot(construction['Time'], construction['Pressure'], linestyle = 'solid', color='green')
plt.xlim(120, 420)
plt.ylim(950, 1050)
plt.title('construction Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Pressure'])

plt.show()
```
