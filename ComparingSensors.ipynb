```python
inside = pd.read_csv('FiveMinutesInside.csv')
inside = inside.iloc[30:301]
```

```python
print("Indoor Mean, Standard Deviation, variance, and uncertainty for other group data")
mean_t = np.mean(inside['Temp'])
std_t = np.std(inside['Temp'])
variancet = np.var(inside['Temp'])
uncertaintyt_inside = (std_t/(270)**.5)
print("Temp", mean_t, std_t, variancet, uncertaintyt_inside)

mean_h = np.mean(inside['Humidity'])
std_h = np.std(inside['Humidity'])
varianceh = np.var(inside['Humidity'])
uncertaintyh_inside = (std_h/(270)**.5)
print("Humidity", mean_h, std_h, varianceh, uncertaintyh_inside)

mean_p = np.mean(inside['Pressure'])
std_p = np.std(inside['Pressure'])
variancep = np.var(inside['Pressure'])
uncertaintyp_inside = (std_p/(270)**.5)
print("Pressure", mean_p, std_p, variancep, uncertaintyp_inside)

mean_a = np.mean(inside['Altitude'])
std_a = np.std(inside['Altitude'])
variancea= np.var(inside['Altitude'])
uncertaintya_inside = (std_a/(270)**.5)
print("Altitude", mean_a, std_a, variancea, uncertaintya_inside)
```

```python
plt.plot(inside['Time'], inside['Temp'], linestyle = 'solid', color='red')
plt.plot(inside['Time'], inside['Humidity'], linestyle = 'solid', color='blue')
#plt.plot(indoor['Time'], indoor['Pressure'], linestyle = 'solid', color='green')
plt.plot(inside['Time'], inside['Altitude'], linestyle = 'solid', color='purple')

plt.title('Indoor Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Temperature', 'Humidity', 'Altitude'])

plt.show()

plt.plot(inside['Time'], inside['Pressure'], linestyle = 'solid', color='green')
plt.title('Indoor Time Series Plot')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Pressure'])

plt.show()
```

```python
print("Time domains don't match")
plt.plot(indoor['Time'], indoor['Temperature'], linestyle = 'solid', color='red')
plt.plot(indoor['Time'], indoor['Humidity'], linestyle = 'solid', color='blue')
plt.plot(indoor['Time'], indoor['Pressure'], linestyle = 'solid', color='green')
plt.plot(indoor['Time'], indoor['Altitude'], linestyle = 'solid', color='purple')


plt.plot(inside['Time'], inside['Temp'], linestyle = 'dashed', color='red')
plt.plot(inside['Time'], inside['Humidity'], linestyle = 'dashed', color='blue')
plt.plot(inside['Time'], inside['Pressure'], linestyle = 'dashed', color='green')
plt.plot(inside['Time'], inside['Altitude'], linestyle = 'dashed', color='purple')


plt.title('Time Series Plot Both Data Sets')
plt.xlabel('Time (seconds)')
plt.ylabel('Data')
plt.legend(['Temperature', 'Humidity', 'Pressure', 'Altitude'])

plt.show()
```

```python
from scipy import stats
plt.hist(indoor["Temperature"],bins=10, color='red')
plt.hist(inside["Temp"],bins=10, color='blue')
plt.legend(['T', 'T Other Data'])

mean_T = np.mean(indoor['Temperature'])
std_T = np.std(indoor['Temperature'])
x = np.linspace(mean_T - 3*std_T, mean_T + 3*std_T, 100)
plt.plot(x, stats.norm.pdf(x, mean_T, std_T))

x = np.linspace(mean_t - 3*std_t, mean_t + 3*std_t, 100)
plt.plot(x, stats.norm.pdf(x, mean_t, std_t))
plt.title('Temperature')
plt.show()


plt.hist(indoor["Humidity"],bins=10, color='red')
plt.hist(inside["Humidity"],bins=10, color='blue')
plt.legend(['H', 'H Other Data'])

mean_H = np.mean(indoor['Humidity'])
std_H = np.std(indoor['Humidity'])
x = np.linspace(mean_H - 3*std_h, mean_H + 3*std_H, 100)
plt.plot(x, stats.norm.pdf(x, mean_H, std_H))

x = np.linspace(mean_h - 3*std_h, mean_h + 3*std_h, 100)
plt.plot(x, stats.norm.pdf(x, mean_h, std_h))
plt.title('Humidity')
plt.show()

plt.hist(indoor["Altitude"],bins=20, color='red')
plt.hist(inside["Altitude"],bins=20,color='blue')
plt.legend(['A', 'A Other Data'])
plt.title('Altitude')

mean_A = np.mean(indoor['Altitude'])
std_A = np.std(indoor['Altitude'])
x = np.linspace(mean_A - 3*std_A, mean_A + 3*std_A, 100)
plt.plot(x, stats.norm.pdf(x, mean_A, std_A))

x = np.linspace(mean_a - 3*std_a, mean_a + 3*std_a, 100)
plt.plot(x, stats.norm.pdf(x, mean_a, std_a))
plt.show()

plt.hist(indoor["Pressure"],bins=10, color='red')
plt.hist(inside["Pressure"],bins=10,color='blue')
plt.title('Pressure')
plt.legend(['P','P Other Data'])

mean_P = np.mean(indoor['Pressure'])
std_P = np.std(indoor['Pressure'])
x = np.linspace(mean_P - 3*std_P, mean_P + 3*std_P, 100)
plt.plot(x, stats.norm.pdf(x, mean_P, std_P))

x = np.linspace(mean_p - 3*std_p, mean_p + 3*std_t, 100)
plt.plot(x, stats.norm.pdf(x, mean_p, std_p))
plt.show() 
```
